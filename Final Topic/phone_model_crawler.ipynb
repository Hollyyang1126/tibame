```python
import requests
import codecs
import csv
import io
from bs4 import BeautifulSoup
from opencc import OpenCC
import pandas as pd
```

```python
page_1 = 1
phone_id_list = []
while True:
    r2=requests.get(
        f"https://www.dcfever.com/phones/brand.php?id=20&page={page_1}",
        headers={
                "User-Agent":"Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.163 Safari/537.36",
                "Referer":"https://www.dcfever.com/phones/secondhand.php?id=280"
            },
        params={
            }
        )

    b2 = BeautifulSoup(r2.text,"html.parser")
    phone_id = b2.find_all("div",{"class":"model_no"})

    for p in phone_id:
        phone_id_list.append(p.text)
    

    page_1 += 1


    if phone_id == []:
        break
# print(phone_id_list)
f = open("phone_model.txt", "a", encoding="utf-8", newline="")
for n in phone_id_list:
    name = n.replace("\n", "")
    name = n.replace(" ", "")
    name = n.lower()
    f.write(name)
    print(name)
f.close()
```
