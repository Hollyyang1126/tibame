```python
import pandas as pd
df = pd.read_csv("./ptt_mi_c_chi.csv")
# fname2 = pd.read_csv("./ptt_iphone._chi.csv")
print("第一個資料", df.shape)
# print("第二個資料", fname2.shape)
```

```python
# 分詞, 數次數, 把慣用係數乘上去, 總分排順序, jieba參考：https://github.com/fxsjy/jieba
# 標題處理, 
import jieba, jieba.analyse
# 載入要得辭典 和不要得辭典
jieba.load_userdict("./mydict.txt")
jieba.analyse.set_stop_words("./stopwords.txt") 
def jana(s):
    s = s.lower()
    s = s.replace("\r", "").replace("\n", "")
    s = s.strip()
    return jieba.analyse.extract_tags(s, 100, withWeight=True)
#keyword = fname["標題"].apply(jana)
#keyword
```

```python
# 內文處理
# keyartical = df["內容"].apply(jana)
# keyartical
```

```python
# 找出內文裡面有 list k 這些字的文章, 有的為 True
def check(s):
    k = ["電量", "充電速度", "使用時間", "qc3.0", "無線充電", "快充", "閃充", "電池", "續航力", "充電"]
    if any(i in s for i in k):
        return s.index
ide = df["內容"].apply(check)
# 把內容裡面有 list k 裡面的文字的留下, 總共有 154筆資料
ide[ide.isna() != 1]
```

```python
# 將有關鍵字的這些文章給一個標籤: 續航力 # 把 true 變成 0, false 變成 1
import numpy as np
k = np.array(ide.isna())
k = k.astype(int)
sum = 0
for i in range(len(k)):
    if k[i] == 0:
        k[i] = 1
        sum +=1
    else:
        k[i] =0
df["續航力"] = k
df
```

```python
# 標籤：相機
def check(s):
    k2 = ["畫素", "廣角", "錄影", "光圈", "多鏡頭", "鏡頭", "對焦", "夜拍", "自拍", "遠景", "近景", "修圖", "變焦", "影像",
          "照片", "相片", "畫面", "影像", "相機", "影相", "單鏡"]
    if any(i in s for i in k2):
        return s.index 
ide = df["內容"].apply(check)
# 把內容裡面有 list k 裡面的文字的留下
#ide[ide.isna() != 1]
k = np.array(ide.isna())
k = k.astype(int)
sum = 0
for i in range(len(k)):
    if k[i] == 0:
        k[i] = 1
        sum +=1
    else:
        k[i] =0
df["相機"] = k
df
```

```python
# 標籤：效能
def check(s):
    k3 = ["處理器", "cpu", "高通", "a13", "a12", "a11", "a10", "多核心", "snapdragon", "klein", "麒麟", "效能",
          "流暢", "穩定","順暢", "軟件", "4g", "5g", "高效能", "硬體", "絕對求生", "吃雞", "高幀數", "lag", "崩壞3rd",
          "傳說對決", "遊戲效能", "遊戲體驗", "rog phone", "ram", "遊戲", "game", "系統"]
    if any(i in s for i in k3):
        return s
ide = df["內容"].apply(check)
k = np.array(ide.isna())
k = k.astype(int)
sum = 0
for i in range(len(k)):
    if k[i] == 0:
        k[i] = 1
        sum +=1
    else:
        k[i] =0
df["效能"] = k
df
```

```python
# 標籤：外觀
def check(s):
    k5 = ["挖屏", "曲面", "大螢幕", "小螢幕", "qhd", "oled", "rgb", "尺寸", "smiled", "螢幕大小", "ips", "機身", 
          "外型", "外形", "screen mon", "介面", "螢幕", "熒幕", "外觀", "外表", "玫瑰金", "夜幕綠", "香檳金", "紫色",
          "綠色", "黃色", "白色", "紅色", "黑色", "太空灰", "灰白色", "曜石黑", "銀色", "銀白色", "金色", "灰色"]
    if any(i in s for i in k5):
        return s.index 
ide = df["內容"].apply(check)
k = np.array(ide.isna())
k = k.astype(int)
sum = 0
for i in range(len(k)):
    if k[i] == 0:
        k[i] = 1
        sum +=1
    else:
        k[i] =0
df["外觀"] = k
df
```
