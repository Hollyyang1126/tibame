```python
import requests
import codecs
import csv
import io
from bs4 import BeautifulSoup
from opencc import OpenCC
import pandas as pd
```

```python
page_1 = 1
phone_id_list = []
while True:
    r2=requests.get(
        f"https://www.dcfever.com/phones/brand.php?id=49&page={page_1}",
        headers={
                "User-Agent":"Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.163 Safari/537.36",
                "Referer":"https://www.dcfever.com/phones/secondhand.php?id=280"
            },
        params={
            }
        )

    b2 = BeautifulSoup(r2.text,"html.parser")
    phone_id = b2.find_all("div",{"class":"image"})

    for p in phone_id:
        str_ = (p.a.attrs["href"]).split("=")
        phone_id_list.append(str_[-1])

    page_1 += 1


    if phone_id == []:
        break
```

```python
with open ("apple.csv", "a", encoding="utf-8", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["型號", "評論"])

for i in phone_id_list:
    page = 1
    while True:
        r3=requests.get(
            f"https://www.dcfever.com/phones/readreviews.php?id={i}&page={page}",
            headers={
                "User-Agent":"Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.163 Safari/537.36",
                "Referer":"https://www.dcfever.com/phones/secondhand.php?id=280"
            }
        )
        b3 = BeautifulSoup(r3.text,"html.parser")
        content_list = []
        contents = b3.find_all("div",{"class":"review_text"})
        for n in contents:
            content_list.append(n.text)

        if len(content_list) == 0:
            break

        page += 1

        name = b3.find("div",{"class":"gadget_model_no"}).text
        name = name.replace("\n", "")
        name = name.replace("\r", "")
        name = name.replace("\t", "")

        with open ("apple.csv", "a", encoding="utf-8", newline="") as file:
            writer = csv.writer(file)
            for n in range(len(content_list)):
                writer.writerow([name, content_list[n]])
```
