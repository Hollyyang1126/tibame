```python
line_list = []
words = ["※", "http"]
for line in open("phone_comment_chi.txt",encoding='utf-8', newline="") :  
    if "※" not in line:
        if "http" not in line:
            line_list.append(line)
line_list
```

```python
comment_list = []
for line in line_list:
    line = line.split(": ")
    line = line[-1].rsplit(" ", 2)
    if line[0] != "" and line[0] != "\n":
        comment_list.append(line[0])
comment_list
```

```python
comments = "\n".join(comment_list)
f = open("phone_comment_v1_chi.txt", "w", encoding = "utf-8")
f.write(comments)
f.close()
```

```python
file1 = open('phone_comment_v1_chi.txt', 'r', encoding='utf-8')
file2 = open('phone_comment_final_chi.txt', 'w', encoding='utf-8')
try:
    for line in file1.readlines():
        if line == '\n':
            line = line.strip("\n")
        file2.write(line)
finally:
    file1.close()
    file2.close()
```
